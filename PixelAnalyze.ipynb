```python
import os
from PIL import Image
import numpy as np
import subprocess
```

```python
dataset_path_root = "Data/medical data/"
datasets = ["melanoma_detection/", "skin_lesions_classification/"]
```

```python
sum_rgb = np.zeros(3)
sum_sq_rgb = np.zeros(3)
num_pixels = 0

results = {}

#Loop trough type folders
for dataset in datasets:
    dataset_path = os.path.join(dataset_path_root, dataset)
    fileNames = []
    imgCount = 0
    for type_folder in os.listdir(dataset_path):
        type_folder_path = os.path.join(dataset_path, type_folder)
        if not os.path.isdir(type_folder_path): continue
        #Loop through class folders
        for class_folder in os.listdir(type_folder_path):
            class_folder_path = os.path.join(type_folder_path, class_folder)
            if not os.path.isdir(class_folder_path): continue
            #Loop through images
            for image_file in os.listdir(class_folder_path):
                image_path = os.path.join(class_folder_path, image_file)
                if not os.path.isfile(image_path): continue
                
                #Open image and convert to numpy array
                with Image.open(image_path) as img:
                    img_array = np.array(img.convert("RGB")) / 255.0
                    print(f"Image: {image_file}, Shape: {img_array.shape}, Dtype: {img_array.dtype}")

                    pixels = img_array.reshape(-1, 3)

                    sum_rgb += pixels.sum(axis=0)
                    sum_sq_rgb += (pixels ** 2).sum(axis=0)
                    num_pixels += pixels.shape[0]
                
                fileNames.append(image_file)
                imgCount += 1
                if imgCount % 100 == 0:
                    print(f"Processed {imgCount} images...")

    mean = sum_rgb / num_pixels
    std = np.sqrt((sum_sq_rgb / num_pixels) - (mean ** 2))


    occurances = {}
    duplicate_files = []
    for fName in fileNames:
        if fName not in occurances:
            occurances[fName] = 0
        occurances[fName] += 1

    for fName, count in occurances.items():
        if count > 1:
            print(f"File: {fName} occures more than once")
            duplicate_files.append(fName)

    
    results[dataset] = {}
    results[dataset]["mean"] = mean
    results[dataset]["std"] = std
    results[dataset]["duplicates"] = duplicate_files
```
