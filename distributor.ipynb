```python
import os
import shutil
import random
```

```python
root = "Data/medical data/modified"
#classname = "nevus"
#classname = "seborrheic_keratosis"
classname = "melanoma"
types = ["train", "test", "val"]
```

```python
#amounts = [1475, 184, 184] #nevus (1843)
amounts = [308, 39, 39] #seborrheic_keratosis (386)


for idx, t in enumerate(types):
    break
    dst = f"{root}/{t}/{classname}"
    os.makedirs(dst, exist_ok=True)
    filenames = os.listdir(f"{root}/{classname}")
    selected = []
    if t == "val":
        selected = filenames
    else:
        selected = random.sample(filenames, amounts[idx])
    for filename in selected:
        shutil.move(f"{root}/{classname}/{filename}", dst)



```
